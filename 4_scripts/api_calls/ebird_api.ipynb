```python
def get_ebird_data(url):

  payload={}
  headers = {
    'X-eBirdApiToken': '49aha8e683li'
  }

  response = requests.request("GET", url, headers=headers, data=payload)

  data = pandas.read_json(response.text)

  return data
```
